```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(2, 2, ".")
```

```python
plt.get_backend()
```

```python
x = np.array([1,2,3,4,5])
y = x*3
plt.plot(x,y)
```

```python
# using artist layer
fig = plt.figure()
ax = plt.subplot()
ax.plot(x, y)

plt.title("basic plot", family="serif", color="#00A3ff", weight="bold", size=20)
plt.xlabel("x", family="serif", color="#00A3ff", weight="bold", size=20)
plt.ylabel("y",family="serif", color="#00A3ff", weight="bold", size=20)
```

```python
fig = plt.figure()
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)
ax1.plot(x, y,".")
ax2.plot(x, y,"-")
ax3.plot(x, y,"o")
ax4.plot(x, y,"x")
```
