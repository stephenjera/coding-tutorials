```python
import requests
import pandas as pd
```

```python
response = requests.get("https://api.sampleapis.com/countries/countries")
data = response.json()
df = pd.DataFrame.from_records(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df2 = pd.json_normalize(df['media'])
```

```python
df2.head()
```

```python
df3 = pd.concat([df,df2], axis=1)
```

```python
df3.head()
```

```python
df3.info()
```

```python
df4 = df3.drop('media', axis=1)
```

```python
df4.head()
```

```python
# Most popular currencies 
df4.groupby('currency').count().sort_values(by='name', ascending=False)['name']
```

```python
# Top 3 countries by population
df4.sort_values(by='population', ascending=False).head(3)
```
