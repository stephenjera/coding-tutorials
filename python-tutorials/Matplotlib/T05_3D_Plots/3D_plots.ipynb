```python
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load the iris dataset from seaborn
iris = sns.load_dataset('iris')
iris[4:10]
```

## Line plot

```python
fig = plt.figure(figsize=(20, 15))
ax = fig.add_subplot(111, projection="3d")

for species in iris["species"].unique():
    iris_species = iris[iris["species"] == species]
    ax.scatter(
        iris_species["sepal_length"],
        iris_species["sepal_width"],
        iris_species["petal_length"],
        label=species,
    )

ax.set_xlabel("Sepal Length", fontsize=14)
ax.set_ylabel("Sepal Width", fontsize=14)
ax.set_zlabel("Petal Length", fontsize=14)
ax.tick_params(axis="both", labelsize=12)
ax.legend(fontsize=12)
plt.show()
```
