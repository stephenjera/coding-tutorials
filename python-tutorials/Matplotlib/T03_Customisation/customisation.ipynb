```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.tail()
```

## Colour Markers and Line Style

```python
# Line plot for sepal length with different flower types
for flower in df["species"].unique():
    plt.plot(df[df["species"] == flower]["sepal_length"], label=flower)
plt.title("Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()
```

## Title, Labels and Legend

```python
# Histogram for petal length with title, labels, and legend
for flower in df["species"].unique():
    plt.hist(df[df["species"] == flower]["petal_length"], label=flower, alpha=0.5)
plt.title("Distribution of Petal Lengths")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Axes and Grid

```python
# Scatter plot for sepal width vs petal width with axes and grid
for flower in df["species"].unique():
    plt.scatter(
        df[df["species"] == flower]["sepal_width"],
        df[df["species"] == flower]["petal_width"],
        label=flower,
    )
plt.title("Sepal Width vs Petal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Petal Width (cm)")
plt.grid(True)
plt.legend()
plt.show()
```

## Text and Annotation

```python
# Line plot for sepal length with text and annotations
for flower in df["species"].unique():
    plt.plot(df[df["species"] == flower]["sepal_length"], label=flower)
plt.title("Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.text(50, 5, "Iris-setosa", fontsize=12)
plt.annotate(
    "Iris-virginica",
    xy=(100, 6),
    xytext=(100, 5),
    arrowprops=dict(facecolor="black", shrink=0.05),
)
plt.legend()
plt.show()
```

## Figure size and DPI

```python
# Bar chart for mean sepal width with figure size and DPI
mean_sepal_width = df.groupby('species')['sepal_width'].mean()
plt.figure(figsize=(8, 6), dpi=80)
plt.bar(mean_sepal_width.index, mean_sepal_width)
plt.title('Mean Sepal Width')
plt.xlabel('Flower')
plt.ylabel('Mean Sepal Width (cm)')
plt.show()
```
