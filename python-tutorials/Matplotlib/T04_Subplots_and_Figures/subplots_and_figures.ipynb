## Creating Subplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the iris dataset from seaborn
iris = sns.load_dataset('iris')
iris[4:10]
```

```python
# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 2)

# Plot sepal length and width on the first subplot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, ax=axs[0, 0])

# Plot petal length and width on the second subplot
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris, ax=axs[0, 1])

# Plot a histogram of sepal length on the third subplot
sns.histplot(iris['sepal_length'], ax=axs[1, 0])

# Plot a histogram of sepal width on the fourth subplot
sns.histplot(iris['sepal_width'], ax=axs[1, 1])

plt.tight_layout()
plt.show()
```

```python
# Create a figure and a set of subplots with shared axes
fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex='col', sharey='row')

# Scatter plot for sepal length vs sepal width on the first subplot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, ax=axs[0, 0])

# Scatter plot for petal length vs sepal width on the second subplot
sns.scatterplot(x='petal_length', y='sepal_width', hue='species', data=iris, ax=axs[0, 1])

# Scatter plot for sepal length vs petal width on the third subplot
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris, ax=axs[1, 0])

# Scatter plot for petal length vs petal width on the fourth subplot
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris, ax=axs[1, 1])

plt.tight_layout()
plt.show()
```

## Insets and Grids of Subplots:

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig, ax = plt.subplots()
ax.plot(iris["sepal_length"])
ax.set_title("Sepal Length")

# Create an inset plot within the main plot
ax_ins = inset_axes(ax, width="50%", height="50%", loc="upper right")
ax_ins.hist(iris["sepal_length"])
ax_ins.set_title("Histogram")

plt.show()
```

## Figures

```python
# First figure
plt.figure(1)
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title("Figure 1: Sepal Length vs Width")

# Second figure
plt.figure(2)
sns.scatterplot(x="petal_length", y="petal_width", hue="species", data=iris)
plt.title("Figure 2: Petal Length vs Width")

plt.show()
```
